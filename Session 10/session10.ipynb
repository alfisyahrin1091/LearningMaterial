```python
var1 = input('Angka Pertama ')
var2 = input('Angka Kedua ')
var3 = input('Operator ')

var4 = int(var1)
var5 = int(var2)


if (var3 =='+'):
  print('hasil : ', (var4+var5))

elif (var3 =='-'):
  print('hasil : ', (var4-var5))

elif (var3 =='*'):
  print('hasil : ', (var4*var5))

elif (var3 =='/'):
  print('hasil : ', (var4/var5))

else:
  print('masukan salah silakan diulangi')
```

# nested if

```python
x = 40

print('program dimulai')

if (x >= 10):
  print('x sama dengan diatas 10')
  if (x > 20):
    print('dan juga diatas 20')
  else:
    print('tapi tidak diatas 20')

else:
  print('x dibawah 10')

print('program selesai')
```
