```python
#Sorted Value

data = {'Nama' : ['Jack' , 'Joko', 'Dwi', 'Andi'],
        'Umur' : [20, 23, 14, 23],
        'Alamat' : ['Surabaya', 'Jakarta', 'Padang', 'Banjarmasin'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Dev', 'Programmer']}

df = pd.DataFrame(data, index = [2,3,0,1])
print(df)
print('')

df_sorted = df.sort_values(by='Nama')
print(df_sorted)
```

```python
#Filtering

data = {'Nama' : ['Jack' , 'Joko', 'Dwi', 'Andi'],
        'Umur' : [20, 23, 34, 23],
        'Alamat' : ['Surabaya', 'Jakarta', 'Padang', 'Banjarmasin'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Developer', 'Programmer']}

df = pd.DataFrame(data)
print(df)
print('')

filtered_df = df[df.Umur > 30]
filtered_df
```

```python
filtered_df = df[(df.Umur > 20) & (df.Alamat == 'Jakarta')]
filtered_df
```

```python
#Using isin()

data = {'Nama' : ['Jack' , 'Joko', 'Dwi', 'Andi'],
        'Umur' : [20, 23, 34, 23],
        'Alamat' : ['Surabaya', 'Jakarta', 'Padang', 'Banjarmasin'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Developer', 'Programmer']}

df = pd.DataFrame(data)
print(df)
print('')

nama =['Dwi', 'Andi']

filtered_df = df[df.Nama.isin(nama)]
filtered_df
```

```python
#Grouping

data = {'Nama' : ['Jack' , 'Joko', 'Dwi', 'Andi'],
        'Umur' : [20, 23, 34, 23],
        'Alamat' : ['Surabaya', 'Jakarta', 'Padang', 'Banjarmasin'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Developer', 'Programmer'],
        'Gaji' : [2000 ,3000, 4000, 5000]}

df = pd.DataFrame(data)
df
```

```python
df.groupby('Pekerjaan').groups
```

```python
df.group = df.groupby('Pekerjaan')

for pekerjaan, group, in df.group:
  print(pekerjaan)
  print(group)
  print('')
```

```python
print(df.group.get_group('DB Admin'))
```

```python
#Aggregate DataFrame

import numpy as np

print('Sebelum Dikelompokkan')

data = {'Nama' : ['Jack' , 'Joko', 'Dwi', 'Andi'],
        'Umur' : [20, 23, 34, 23],
        'Alamat' : ['Surabaya', 'Jakarta', 'Padang', 'Banjarmasin'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Developer', 'Programmer'],
        'Gaji' : [2000 ,3000, 4000, 5000]}

df = pd.DataFrame(data)
print(df)
print('')
print('Sesudah Dikelompokkan')

df_group = df.groupby('Pekerjaan').agg(np.mean)
df_group
```

```python
df_group = df.groupby('Pekerjaan').agg([np.mean, np.sum, np.std])
df_group
```

```python
#Merging DataFrame

import pandas as pd

kiri = pd.DataFrame({'Id' : [1,2,3],
                     'Nama' : ['Andi', 'Joko', 'Budi']})
kanan = pd.DataFrame({'Id' : [1,2,3],
                      'Nama' : ['Mike', 'Jack', 'Jane']})

gabung = pd.merge (kiri, kanan, on='Id')
gabung 
```

```python
import pandas as pd

karyawan = pd.DataFrame({'Id' : [1,2,3],
                     'Nama' : ['Andi', 'Joko', 'Budi'],
                     'Id Pekerjaan' : [1,2,3]})
pekerjaan = pd.DataFrame({'Id Pekerjaan' : [1,2,3,4,5],
                      'Pekerjaan' : ['DB Admin', 'Programmer', 'Sekretaris', 'OB','Accountant'],
                      'Gaji' : [1000,2000,3000,4000,5000]})

print(karyawan)
print(' ')
print(pekerjaan)
print(' ')


gabung = pd.merge (karyawan, pekerjaan, on='Id Pekerjaan', how = 'left')
gabung 
```
