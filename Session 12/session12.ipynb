#accessing with loop

```python
list1 = ['John', 'Paul', 'George', 'Ringo']
list2 = [3.14, True, 8, 'Hello', 'Word']

for i in list1:
  print(i)

print('-----------')

for j in list2:
  print(j, end=' ')
```

```python
list1 = ['John', 'Paul', 'George', 'Ringo']
list2 = [3.14, True, 8, 'Hello', 'Word']

for i in range (len(list1)):
  print(list1[i])

print('------')

for j in range(len(list2)):
  print(list2[j])
```

#Nested List

```python
a = [1, 2, 3, 4]
b = ['a', 'b', 'c']

c = ['hello', a, 'world', b]
print(c)

print(c)
print(c[3])
print(c[3][2])

print('-------')

for i in c:
  print(i)

print('-------')

for j in range(len(c)):
  print(j, end=' ')
```

#Dictionary

```python
cities = {'Indonesia' : 'Jakarta', 
          'Malaysia' : 'Kuala Lumpur',
          'Vietnam' : 'Hanoi',
          'Thailand' : 'Bangkok',
          'Filipina' : 'Manila'
}

print(cities)
print(cities['Vietnam'])
print(cities.get('Thailand'))
```

```python
cities = {'Indonesia' : 'Jakarta', 
          'Malaysia' : 'Kuala Lumpur',
          'Vietnam' : 'Hanoi',
          'Thailand' : 'Bangkok',
          'Filipina' : 'Manila'
}

for i in cities:
  print(i)

print('--------------')

for i in cities:
  print(i, ' : ', cities[i])
```
