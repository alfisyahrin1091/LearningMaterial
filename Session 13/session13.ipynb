```python
#Arbitrary Arguments (args)

def print_mhs(*inp_mhs):
  print('mahasiswa nomor urut kedua adalah :', inp_mhs[1])

print_mhs('andi', 'joko', 'mike', 'joy', 'budi')
```

```python
def print_mhs(inp1, *inp_mhs):
  print('mahasiswa :', inp1)
  for i in inp_mhs:
    print('mahasiswa itu : ', i)

print_mhs('andi', 'joko', 'mike', 'joy', 'budi')
```

```python
#Keyword Arguments

def print_mhs(nama, alamat, umur):
  print('nama : ', nama)
  print('alamat : ', alamat)
  print('umur :', umur)

print_mhs(alamat = 'jakarta', nama = 'andi', umur = '19')
```

```python
#Keyword Arbitrary Arguments (kwargs)

def print_mhs(**mhs):
  print('nama ', mhs['nama'])
  print('alamat ', mhs['alamat'])
  print('umur ', mhs['umur'])

print_mhs(nama='mike', alamat='surabaya', umur=22, jurusan='biologi', nim=1001, semester='5')
```

```python
#Default Parameter Value

def nama_jurusan(jurusan='biologi'):
  print('jurusan mahasiswa', jurusan)

nama_jurusan()
nama_jurusan('matematika')
```

```python
def nama_jurusan(nama, alamat, jurusan='biologi'):
  print('jurusan mahasiswa', jurusan)
  print('nama ', nama)
  print('alamat ', alamat)

nama_jurusan(nama='steve', alamat='banten')
nama_jurusan('gilar', 'palembang', 'matematika')
```

```python
# Passing a List as a Argument

def print_matkul(matkul):
  for i in matkul:
    print('mata kuliah adalah :', i)

mk = ['fisika', 'kimia', 'biologi', 'matematika']
print_matkul(mk)
```

```python
#Return Values

def penjumlahan(a,b):
  c = a+b
  return c

a1 = 5
b1 = 12

c1 = penjumlahan(a1,b1)
print('hasil jumlah', c1)
```

```python
def penjumlahan_pengurangan(a,b):
  c = a+b
  d = a-b
  return c, d

a1 = 5
b1 = 12

c1, c2 = penjumlahan_pengurangan(a1,b1)

print('hasil jumlah' , c1)
print('hasil pengurangan ', c2)
```

```python
def operasi(a2, b2, op):
  if (op == '+'):
    f = a2 + b2
    g = a2 - b2
    return f, op, g

a2= 2
b2= 4

f, op, g = operasi (a2,b2)

print(f)
print(g)
```

```python
#Lambda

x = lambda a, b : a*b

y = x(5,4)
print(y)
```

```python
status = lambda umur : 'remaja' if umur > 10 and umur < 20 else 'bukan remaja'
print(status(25))
```
