# while loop

```python
count = 0
print('program dimulai')
while (count <= 10):
  print('count adalah : ', count)
  count  += 1 
  # count = count + 1

print('program selesai')
```

```python
count = 10
print('program dimulai')
while (count >= 0):
  print(count, end=' ')
  count  -= 1 
  # count = count - 1

print('program selesai')
```

# for loop

```python
print('program dimulai')

for i in range (1, 10, 2):
  print(i)

print('program selesai')  
```

```python
for i in range(20, 0, -2):
  print(i, end=' ')
```

#break loop statement

```python
print('using break inside for loop')

num = 3
for i in range (0, 10):
  if (i==num):
    break
  print(i, end=' ')
  print('berhitung')
  print('hasil kuadrat dari ', i, 'adalah', i**2)

print('end program')
```

```python
i = 0
num = 3

while (i < 10):
  print(i)
  if (i == num):
    break
  
  
  i +=1
```

# Continue Loop Statement

```python
num = 3
for i in range(0,10):
  if (i == num):
    continue

  print(i, end=' ')
  print('contoh continue')

print(' ')
print('program selesai')
```

```python
i = 10
num = 7

while (i > 0):
  i = i - 1
  if i == num:
    continue

  print(i)
```

# for loop with else

```python
for i in range(0,15):
  print(i)
else:
  print('program selesai')

print('out of block')
```

```python
for i in range(0,15):
  print(i)
  if (i == 5):
    break
else:
  print('program selesai')

print('out of block')
```

# Nested Loops

```python
for i in range (0,10):
  for j in range (0,3):
    j +=1
    print(j, end=' ')
  print('i: ', i)

  print ('')
```

```python
for i in range (0,10):

  for j in range (0,3):
    print(i,', ', j, end= '   ')

  print ('')
```
